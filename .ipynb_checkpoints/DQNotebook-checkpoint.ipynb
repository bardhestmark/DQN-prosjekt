```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 50)

lr5e5 = pd.read_csv('./atari_dqn/trained_models/Pong-v0_2021-11-17-120751/Pong-v0.csv', skiprows=1)
lr5e4 = pd.read_csv('./atari_dqn/trained_models/Pong-v0_2021-11-17-120818/Pong-v0.csv', skiprows=1)

print(lr5e5.head())
print(lr5e4.head())

lr5e5.plot(title='Mean reward for LR=5e-5', x='step', y='rew_mean')
lr5e4.plot(title='Mean reward for LR=5e-4', x='step', y='rew_mean')
```
